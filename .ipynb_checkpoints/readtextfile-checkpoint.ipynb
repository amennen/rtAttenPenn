```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import r2_score
import pickle
import pdb
import sklearn
import scipy
import scipy.io
from sklearn import metrics
import os,glob
from matplotlib.pyplot import cm 
import matplotlib
matplotlib.rcParams.update({'font.size': 22})
import pandas as pd
import itertools
import seaborn as sns
from scipy import stats
# specify now which computer you're using!
datapath = '/Users/amennen/Downloads/Libkuman-BRM-2007/'
```

```python
file = open(datapath + 'Libkuman(2007)APP-D.txt','r') 
```

```python
allImages = {}
iter = 0
for line in file:
    #print(line.split())
    if iter>=4:
        #print(line.split())
        data = line.split()
        if len(data) > 0: # because of the last line
            id = data[1]
            if id.isdigit():
                print(id)
                happy = data[3]
                sad = data[7]
                fear = data[13]
                allImages[id] = np.array([float(happy), float(sad), float(fear)])
            else:
                id = data[2]
                happy = data[4]
                sad = data[8]
                fear = data[14]
                allImages[id] = np.array([float(happy), float(sad), float(fear)])
    iter += 1
```
